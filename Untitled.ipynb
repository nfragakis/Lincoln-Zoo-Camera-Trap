```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('updated_2018_detections.csv')
```

```python
import matplotlib.pyplot as plt
sns.set_style('darkgrid')

fig, ax = plt.subplots(figsize=(8, 6))
sns.countplot(x='ShortName', 
              data=df, 
              order=df['ShortName'].value_counts().index)
plt.title('')
plt.xticks(rotation=45);
```
